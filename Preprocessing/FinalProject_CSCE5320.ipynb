```python
# Step 3: Standardize column names
air_quality.columns = air_quality.columns.str.strip().str.lower().str.replace(" ", "_")
weather.columns = weather.columns.str.strip().str.lower().str.replace(" ", "_")
```

```python
air_quality.head(1)
```

```python
weather.head(1)
```

```python
# Step 4: Convert date columns
air_quality["date"] = pd.to_datetime(air_quality["date"], errors="coerce")
weather["last_updated"] = pd.to_datetime(weather["last_updated"], errors="coerce")
weather["date"] = pd.to_datetime(weather["last_updated"].dt.date)
```

```python
air_quality.head(1)
```

```python
weather.head(1)
```

```python
# Step 5: Merge datasets on country, city/location_name, and date
merged = pd.merge(
    air_quality,
    weather,
    left_on=["country", "city", "date"],
    right_on=["country", "location_name", "date"],
    how="inner"
)
```

```python
merged.head(5)
```

```python
# Step 6: Drop irrelevant columns
columns_to_drop = [
    "last_updated", "location_name", "sunrise", "sunset",
    "moonrise", "moonset", "moon_phase", "moon_illumination",
    "localtime", "localtime_epoch", "timezone_id", "utc_offset"
]
merged.drop(columns=[col for col in columns_to_drop if col in merged.columns], inplace=True)
```

```python
merged.head(5)
```

```python
# Step 7: Save cleaned merged dataset
merged.to_csv("merged_cleaned_air_quality_weather.csv", index=False)
files.download("merged_cleaned_air_quality_weather.csv")
```

**How do weather variables (temperature, humidity, wind speed) affect pollution levels across different seasons?**

```python
# Step 2: Import required libraries
import pandas as pd
import plotly.express as px
```

```python
# Step 3: Prepare the data

# Create a season column based on the month
def assign_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

merged['Season'] = merged['date'].dt.month.apply(assign_season)
```
